## iki grafiği yan yana çizmek istersek o zaman matplotlib object oriented yöntemini kullanmalıyız

```python
fig, hg = plt.subplots(1, 2, figsize=(10,5)) # 1,2 : 1 satır 2 sütun 
sns.scatterplot(data=tips, x="total_bill", y="tip", style="sex", hue="sex", ax=hg[0]);
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", ax=hg[1]);

# birincinin ax'ını hg[0] olarak yani satırın ilk indeksi olarak,
# ikincinin ax'ını hg[1] olarak yani satırın ikinci indeksi olarak tanımladık

# subplotlarda sns satırının içinde ax'i tanımlıyoruz!
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=tips, x="total_bill", y="tip", style="sex", hue="sex", alpha=0.7);
```

```python
# s: noktlaların size'ı. nokta büyüklüğünü ayarlıyor. size'la karıştırma!

plt.figure(figsize=(10,5))
sns.scatterplot(data=tips, x="total_bill", y="tip", style="sex", hue="sex", alpha=0., s=300);
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=tips, x="total_bill", y="tip", size="sex", hue="smoker", palette='plasma');

# total_bill ve tip olarak 2 boyutum vardı. size = sex ve hue = smoker olarak 2 boyut daha kattım.
```

**Histogram grafiği için:  Displot ve Histplot**

```python
sns.displot(data=mpg, x="weight");
```

```python
sns.histplot(data=mpg, x="weight");
```

```python
fig, ax= plt.subplots(1,2,figsize=(10,5))

sns.scatterplot(data=tips, x='total_bill',y='tip', s=300, hue='sex', ax=ax[0])

sns.displot(data=mpg, x='weight', ax=ax[1]);

# ax[1] olarak tanımladığım displotu çizemedi. çünkü zaten o bir canvastır, onu bir axes'e atayamazsın.
```

```python
fig, ax= plt.subplots(1,2,figsize=(10,5))
sns.scatterplot(data=tips, x='total_bill',y='tip', s=300, hue='sex', ax=ax[0])
sns.histplot(data=mpg, x='weight', ax=ax[1]);
# sns.displot(data=mpg, x='weight', ax=ax[1]);
```

```python
sns.displot(data=mpg, x="weight", kind="hist");   # zaten default olarak hist çizer. bunu yazmasaydık da histogram çizecekti.
sns.displot(data=mpg, x="weight", kind="kde");

# kde grfiğinin altında kalan kapalı alan her zaman 1'e eşittir.
# kde grafiği bir ihtimaller yüzdesi verir.
```

```python
sns.displot(data=mpg, x="weight", kde=True);

# kde=True dediğimde default olan histogram ile birlikte kde'yi de üzerine çizer.

# kde sayesinde datasetim right skewed mı? left skewed mı görebiliyoruz
```

```python
sns.displot(data=mpg, x="weight", kde = True, aspect=1.5);

# displot'ta boyutu aspect ile ayarlayabiliyoruz.

# figsize şeklinde matplotlib'in object oriented yöntemlerini displotta kullanmıyoruz!
```

**joinplot**

```python
sns.jointplot(data=tips, x="total_bill", y="tip", kind="scatter");  # default olarak scatter çizer.


# kind : { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }  default : scatter
```

```python
sns.jointplot(data=tips, x="total_bill", y="tip", kind="hex");

# koyu renkli yerlerde güçlü bir korelasyon olduğunu, orada verilerin yoğunlaştığını görebiliyoruz
```

```python
sns.jointplot(data=tips, x="total_bill", y="tip", kind="kde");
```

```python
sns.jointplot(data=mpg, x="horsepower", y="acceleration" , kind="scatter");
```
