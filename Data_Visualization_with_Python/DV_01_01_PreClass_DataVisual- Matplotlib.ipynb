```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x,y)

axes[0][1].plot(x,y)

axes[1][0].plot(x,y)
axes[1][0].set_ylabel('Y label')

axes[1][1].plot(a,b)
axes[1][1].set_title('SET Title')

fig.subplots_adjust(wspace=1, hspace=1)

# set_ylabel ile o axes[1][0] axesinin y eksenine etiket ismi verdik.
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x,y)

axes[0][1].plot(x,y)

axes[1][0].plot(x,y)
axes[1][0].set_ylabel('Y label')
axes[1][0].set_xlim(2,6)
axes[1][0].set_ylim(0, 100)

axes[1][1].plot(a,b)
axes[1][1].set_title('TITLE 1,1')

fig.subplots_adjust(wspace=1, hspace=1)

# set_xlim() ve set_ylim() ile axes[1][0] axes'in x ve y limit aralıklarını belirledik.
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x,y)

axes[0][1].plot(x,y)

axes[1][0].plot(x,y)
axes[1][0].set_ylabel('Y label')
axes[1][0].set_xlim(2,6)
axes[1][0].set_ylim(0, 100)

axes[1][1].plot(a,b)
axes[1][1].set_title('TITLE 1,1')

fig.suptitle('Figure Level', fontsize=16)

plt.tight_layout

# fig.suptitle() ile Figure'a super title verdik. fontsize parametresi ile title fontunu 16ya ayarladık.
```

```python
# figure'u kaydedelim.

fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x,y)

axes[0][1].plot(x,y)

axes[1][0].plot(x,y)
axes[1][0].set_ylabel('Y label')
axes[1][0].set_xlim(2,6)
axes[1][0].set_ylim(0, 100)

axes[1][1].plot(a,b)
axes[1][1].set_title('TITLE 1,1')

fig.suptitle('Figure Level', fontsize=16)

plt.tight_layout

fig.savefig('new_subplots.png', bbox_inches='tight')

# figure'ı new_subplots.png adıyla kaydettik.

"""
bbox_inches : str or `.Bbox`, default: :rc:`savefig.bbox`
    Bounding box in inches: only the given portion of the figure is
    saved.  If 'tight', try to figure out the tight bbox of the figure.
"""
```

```python
# figure genişliğini ayarlayalım:

fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x,y)

axes[0][1].plot(x,y)

axes[1][0].plot(x,y)
axes[1][0].set_ylabel('Y label')
axes[1][0].set_xlim(2,6)
axes[1][0].set_ylim(0, 100)

axes[1][1].plot(a,b)
axes[1][1].set_title('TITLE 1,1')

fig.suptitle('Figure Level', fontsize=16)

plt.tight_layout

fig.set_figwidth(10)   # figure genişliğini 10'a çıkarttık

fig.savefig('new_subplots.png', bbox_inches='tight')
```

```python
# figure genişliğini başlangıçta objectleri oluşturuken figsize() parametresi ile ayarlayabiliriz.

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(4,10), dpi = 200)

axes[0][0].plot(x,y)

axes[0][1].plot(x,y)

axes[1][0].plot(x,y)
axes[1][0].set_ylabel('Y label')
axes[1][0].set_xlim(2,6)
axes[1][0].set_ylim(0, 100)

axes[1][1].plot(a,b)
axes[1][1].set_title('TITLE 1,1')

fig.suptitle('Figure Level', fontsize=16)

plt.tight_layout

fig.savefig('new_subplots.png', bbox_inches='tight')
```

## MATPLOTLIB STYLING - LEGENDS

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 11, 10)
x
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x)

ax.plot(x, x**2)

# aynı axes üzerinde iki plotlama tanımaldım. her birini otomatik olarak farklı renlerde çizdi.
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, label = 'X vs X')  

ax.plot(x, x**2, label = 'X vs X^2')

ax.legend()

# her iki plota da labellar ekledim. ve bunları legend komutu ile sol üstte gösterdim.
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, label = 'X vs X')  

ax.plot(x, x**2, label = 'X vs X^2')

ax.legend(loc='lower left')

# legendin konumunu loc ile sol alt yaptım.
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, label = 'X vs X')  

ax.plot(x, x**2, label = 'X vs X^2')

ax.legend(loc=(1.1, 0.5))

# legend'in konumunu bu şekilde de spesific olarak belirleyebiliyoruz
# y ekseninin 1.1 birim sağına kayık, x ekseninin 0.5 birim üzerinde
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, label = 'X vs X')  

ax.plot(x, x**2, label = 'X vs X^2')

ax.legend(loc=(0.5, 0.5))
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, label = 'X vs X')  

ax.plot(x, x**2, label = 'X vs X^2')

ax.legend(loc=(-0.5, 0.5))

# negatif değer ile y eksinin soluna çektik.
```
