```python
def quality(total_bill, tip) :
    if tip / total_bill > 0.25:
        return "Generous"
    else : 
        return "Other"
    
# bırakılan bahşişin toplam hesaba oranı 0.25'ten büyükse bunu Generous olarak, 
# 0.25'ten küçükse bunu Other olarak classify yapmak istiyoruz. 
# kolaylık olsun diye fonksiyonun parametrelerini kullanacağımız sütunların ismileri olarak seçtik.
```

```python
quality(16.99, 1.01)
```

```python
df[['total_bill', 'tip']].apply(lambda df : quality(df['total_bill'], df['tip']), axis = 1)

# df nin total_bill ve tip sütunlarınını quality fonksiyonunun parametreleri olarak 
# fonksiyona tatbik ederek bunu lambda fonksiyonuna soktuk ve bunu bu iki sütuna apply ettik.
# sütunlarda çalıştığımız için axis'i 1 girdik.

# lambda kullanmasaydık quality fonkiyonunun parametreleri yalnızca birer değer alabilecekti.
# bu da hataya neden olacaktı. lambda fonksiyonu ile df'nin belirtilen iki sütununun elemanları 
# sırasıyla lambdadaki işleme girecek.
```

```python
df['Quality'] = df[['total_bill', 'tip']].apply(lambda df : quality(df['total_bill'], df['tip']), axis = 1)

# yukardaki sonucu 'Quality' isimli bir sütun haline getirdik.
```

```python
df
```

```python
# aynı sonucu np.vectorize ile de elde edebiliriz.

df['Quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])

# quality fonksiyonunu numpy'ın vectorize fonksiyonunu kullanarak vektörize ettik.

"""Define a vectorized function which takes a nested sequence of objects or
numpy arrays as inputs and returns a single numpy array or a tuple of numpy
arrays. The vectorized function evaluates `pyfunc` over successive tuples
of the input arrays like the python map function, except it uses the
broadcasting rules of numpy.

The data type of the output of `vectorized` is determined by calling
the function with the first element of the input.  This can be avoided
by specifying the `otypes` argument.
"""
```

## Groupby Operations - Part1

```python
"""
we need to choose a categorical column to call with groupby()

categorical columns are non-continuous

keep in mind, they can still be numerical, such as cabin class categories on a ship
(e.g. Class1, Class2, Class3)

Note that in pandas, calling groupby() by itself creates a "lazy" groupby object
waiting to be evaluated by an aggregate method call.

"""
```

```python
df = pd.read_csv('mpg.csv')
```

```python
df.head()
```

```python
df['model_year'].unique()

# model yıllarının unique değerlerini görelim. hangi modeller var bilmek için
```

```python
df['model_year'].value_counts()

# model_year'a göre datasetimde kaçar value var.
```

```python
df.groupby('model_year')

# veriyi model yıllarına göre grupladık
# bunun bir groupby object olduğuna dikkat et.

# groupby lazy bir değer. çünkü groupby yaptığında henüz herhangi bir işlem yapılmamıştır.
# çünkü o hala aggregate metodtan gelecek hesaplamaları gerçekleştirmeyi bekliyor.
```

```python
df.groupby('model_year').mean()

# model_year şimdi bir index haline geldi. o artık bir column ismi değil.
```

## Missing Data

```python
# when reading in missing values, pandas will display them as NaN values.
```

```python
# options for missing data:
  # keep it
  # remove it
  # replace it
    
# there is never 100% correct approach that applies to all circumstances, 
# it all depends on the exact situation you encounter!
```

```python
# keeping the missing data
  # PROS:
    # easiest to do
    # does not manipulate or change the true data
  # CONS:
    # many methods do not support NaN
    # ofter there are reasonable guesses
```

```python
# Dropping or removing the missing data
  # PROS:
    # easy to do
    # can be based on rules
  # CONS:
    # potential to lose a lot of data or usefull information
    # limits trained models for future data.
```

```python
# Filling in the missing data
  # PROS:
    # potential to save a lot of data for use in training a model
  # CONS:
    # hardest to do and somewhat arbitrary
    # potential to lead to false conclusions
  # fill with same value
    # good choice if NaN was a placeholder
  # fill with interpolated or estimated value
    # much harder and requires reasonable assumptions
```

```python
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})
```

## how = left merge

```python
"""
how : {'left', 'right', 'outer', 'inner', 'cross'}, default 'inner'
    Type of merge to be performed.

    * left: use only keys from left frame, similar to a SQL left outer join;
      preserve key order.
    * right: use only keys from right frame, similar to a SQL right outer join;
      preserve key order.
    * outer: use union of keys from both frames, similar to a SQL full outer
      join; sort keys lexicographically.
    * inner: use intersection of keys from both frames, similar to a SQL inner
      join; preserve the order of the left keys.
    * cross: creates the cartesian product from both frames, preserves the order
      of the left keys.
"""
```

```python
# how = 'left' olarak merge ettiğimizde: 

# her iki tablo da birleştirilecek ancak sonuç soldaki tablounun kayıtlarından oluşacak, 
# bu kayıtlardan sağdaki tablo olmayan data yerine NaN atanacaktır.
```

```python
pd.merge(left=registrations, right=logins, how='left', on='name')

# parantez içinde sol ve sağ tabloların hangileri olduğunu belirtiyoruz!

# soldaki tablonun name'lerini olduğu gibi aldı. sağdaki tabloda bu verilere karşılık boş olan data yerine NaN atadı
```
